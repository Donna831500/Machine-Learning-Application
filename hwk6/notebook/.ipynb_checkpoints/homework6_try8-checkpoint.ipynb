```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix,roc_curve,auc
import math
from IPython.display import display
import warnings
import statistics
warnings.filterwarnings('ignore')
import sklearn.model_selection as model_selection
from sklearn import svm
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.svm import LinearSVC
from sklearn import preprocessing
from sklearn.pipeline import Pipeline
```

```python
def convert_label(s):
    if s=='B':
        return 0;
    elif s=='M':
        return 1;
    
    
def get_mode(input_list):
    number_of_one = input_list.count(1)
    number_of_zero = input_list.count(0)
    if number_of_one>=number_of_zero:
        return 1;
    else:
        return 0;
    
    
def contain_both_class(input_df):
    Y_train_list = input_df[4].tolist()
    number_of_one = Y_train_list.count(1)
    number_of_zero = Y_train_list.count(0)
    if number_of_one>0 and number_of_zero>0:
        return True
    else:
        return False
```

```python
temp = pd.read_csv('../data/wdbc.csv',
                na_values='?',
                header=None,
                index_col=None)
df = temp.iloc[:,1:]
df['label'] = df.iloc[:, 0].apply(convert_label)
df = df.iloc[:,1:]
df_X = df.iloc[:,:-1]
df_Y = df.iloc[:,-1:]
df_B = df[df.iloc[:, -1] == 0]
df_M = df[df.iloc[:, -1] == 1]
```

<h5>i)<h5>

```python
temp_list = list(range(-3,6))
c_list = []
for i in temp_list:
    c_list.append(pow(10,i))
```

```python
df_B_X = df_B.iloc[:,:-1]
df_B_Y = df_B.iloc[:,-1:]
df_M_X = df_M.iloc[:,:-1]
df_M_Y = df_M.iloc[:,-1:]
```

```python
accuracy_list_test = []
precision_list_test = []
recall_list_test = []
F1_score_list_test = []
auc_list_test = []
accuracy_list_train = []
precision_list_train = []
recall_list_train = []
F1_score_list_train = []
auc_list_train = []

for M in range(0,30):
    rs = M
    X_B_train, X_B_test, Y_B_train, Y_B_test = model_selection.train_test_split(df_B_X, df_B_Y, train_size=0.8,test_size=0.2, random_state=rs);
    X_M_train, X_M_test, Y_M_train, Y_M_test = model_selection.train_test_split(df_M_X, df_M_Y, train_size=0.8,test_size=0.2, random_state=rs);
    #X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X, Y, train_size=0.7,test_size=0.3, random_state=10);
    X_train = pd.concat([X_B_train,X_M_train])
    X_test = pd.concat([X_B_test,X_M_test])
    Y_train = pd.concat([Y_B_train,Y_M_train])
    Y_test = pd.concat([Y_B_test,Y_M_test])
    
    param_grid = {
        'svc__penalty':['l1'],
        'svc__C':c_list,
        'svc__dual':[False]
    }
    pipe = Pipeline(steps=[('scaler',preprocessing.MinMaxScaler()),('svc', LinearSVC())])
    clf = GridSearchCV(estimator = pipe, n_jobs=-1, param_grid=param_grid,cv=5,scoring='accuracy')
    clf.fit(X_train, Y_train)
 
    # Train
    Y_predict_train = clf.predict(X_train)
    tn, fp, fn, tp = confusion_matrix(y_true = Y_train, y_pred = Y_predict_train).ravel()
    accuracy = (tn+tp)/(tn+fp+fn+tp)
    precision = tp/(tp+fp)
    recall = tp/(tp+fn)
    f1_score = 2*((precision*recall)/(precision+recall))
    fpr, tpr, thresholds = roc_curve(y_true = Y_train,y_score=Y_predict_train,pos_label=1)
    area_under_curve = auc(fpr, tpr)
    accuracy_list_train.append(accuracy)
    precision_list_train.append(precision)
    recall_list_train.append(recall)
    F1_score_list_train.append(f1_score)
    auc_list_train.append(area_under_curve)

    # Test
    Y_predict_test = clf.predict(X_test)
    tn, fp, fn, tp = confusion_matrix(y_true = Y_test, y_pred = Y_predict_test).ravel()
    accuracy = (tn+tp)/(tn+fp+fn+tp)
    precision = tp/(tp+fp)
    recall = tp/(tp+fn)
    f1_score = 2*((precision*recall)/(precision+recall))
    fpr, tpr, thresholds = roc_curve(y_true = Y_test,y_score=Y_predict_test,pos_label=1)
    area_under_curve = auc(fpr, tpr)
    accuracy_list_test.append(accuracy)
    precision_list_test.append(precision)
    recall_list_test.append(recall)
    F1_score_list_test.append(f1_score)
    auc_list_test.append(area_under_curve)
```

```python
accuracy_list_test = []
precision_list_test = []
recall_list_test = []
F1_score_list_test = []
auc_list_test = []
accuracy_list_train = []
precision_list_train = []
recall_list_train = []
F1_score_list_train = []
auc_list_train = []

for M in range(0,30):

    rs = M
    X_B_train, X_B_test, Y_B_train, Y_B_test = model_selection.train_test_split(df_B_X, df_B_Y, train_size=0.8,test_size=0.2, random_state=rs);
    X_M_train, X_M_test, Y_M_train, Y_M_test = model_selection.train_test_split(df_M_X, df_M_Y, train_size=0.8,test_size=0.2, random_state=rs);
    #X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X, Y, train_size=0.7,test_size=0.3, random_state=10);
    X_test = pd.concat([X_B_test,X_M_test])
    Y_test = pd.concat([Y_B_test,Y_M_test])
    X_train = pd.concat([X_B_train,X_M_train])
    Y_train = pd.concat([Y_B_train,Y_M_train])

    label_B_X, unlabel_B_X, label_B_Y, unlabel_B_Y = model_selection.train_test_split(X_B_train, Y_B_train, train_size=0.5,test_size=0.5, random_state=rs);
    label_M_X, unlabel_M_X, label_M_Y, unlabel_M_Y = model_selection.train_test_split(X_M_train, Y_M_train, train_size=0.5,test_size=0.5, random_state=rs);
    label_X = pd.concat([label_B_X,label_M_X])
    unlabel_X = pd.concat([unlabel_B_X,unlabel_M_X])
    label_Y = pd.concat([label_B_Y,label_M_Y])
    unlabel_Y = pd.concat([unlabel_B_Y,unlabel_M_Y])

    # get best C by CV
    param_grid = {
        'svc__penalty':['l1'],
        'svc__C':c_list,
        'svc__dual':[False]
    }
    pipe = Pipeline(steps=[('scaler',preprocessing.MinMaxScaler()),('svc', LinearSVC())])
    clf = GridSearchCV(estimator = pipe, n_jobs=-1, param_grid=param_grid,cv=5,scoring='accuracy')
    clf.fit(label_X, label_Y)
    best_C = clf.best_params_.get('svc__C')
    
    # find and sort distance
    min_max_scaler = preprocessing.MinMaxScaler()
    x_scaled = min_max_scaler.fit_transform(label_X)
    label_X_normalize = pd.DataFrame(x_scaled)
    
    lsvc = LinearSVC(penalty='l1',C=best_C,dual=False)
    lsvc.fit(label_X_normalize,label_Y)
    y = lsvc.decision_function(unlabel_X)
    w_norm = np.linalg.norm(lsvc.coef_)
    distance_list = list(y / w_norm)
    raw_distance = list(abs(clf.decision_function(unlabel_X)))
    adding_list = sorted(range(len(raw_distance)), key=lambda k: raw_distance[k],reverse=True)
    
    # repeatly add data point
    for index in range(0,len(adding_list)):
        current_data = pd.DataFrame(unlabel_X.iloc[adding_list[index],:]).transpose()

        #---------normalize data---------
        temp = pd.concat([label_X,current_data])
        min_max_scaler = preprocessing.MinMaxScaler()
        x_scaled = min_max_scaler.fit_transform(temp)
        temp_normalize = pd.DataFrame(x_scaled)
        label_X_normalize = temp_normalize.iloc[:-1,:]
        current_data_normalize = temp_normalize.iloc[[-1]]

        lsvc = LinearSVC(penalty='l1',C=best_C,dual=False)
        lsvc.fit(label_X_normalize,label_Y)
        current_y = pd.DataFrame(data=lsvc.predict(current_data_normalize),columns=['label'])
        label_X = pd.concat([label_X,current_data])
        label_Y = pd.concat([label_Y,current_y])

    min_max_scaler = preprocessing.MinMaxScaler()
    x_scaled = min_max_scaler.fit_transform(label_X)
    label_X_normalize = pd.DataFrame(x_scaled)
    lsvc = LinearSVC(penalty='l1',C=best_C,dual=False)
    lsvc.fit(label_X_normalize,label_Y)   
    
    # normalize train data + predict
    min_max_scaler = preprocessing.MinMaxScaler()
    x_scaled = min_max_scaler.fit_transform(X_train)
    X_train_normalize = pd.DataFrame(x_scaled)
    Y_predict_train = lsvc.predict(X_train_normalize)

    # normalize test data + predict
    min_max_scaler = preprocessing.MinMaxScaler()
    x_scaled = min_max_scaler.fit_transform(X_test)
    X_test_normalize = pd.DataFrame(x_scaled)
    Y_predict_test = lsvc.predict(X_test_normalize)
    
    # Train
    tn, fp, fn, tp = confusion_matrix(y_true = Y_train, y_pred = Y_predict_train).ravel()
    accuracy = (tn+tp)/(tn+fp+fn+tp)
    precision = tp/(tp+fp)
    recall = tp/(tp+fn)
    f1_score = 2*((precision*recall)/(precision+recall))
    fpr, tpr, thresholds = roc_curve(y_true = Y_train,y_score=Y_predict_train,pos_label=1)
    area_under_curve = auc(fpr, tpr)
    accuracy_list_train.append(accuracy)
    precision_list_train.append(precision)
    recall_list_train.append(recall)
    F1_score_list_train.append(f1_score)
    auc_list_train.append(area_under_curve)

    # Test
    tn, fp, fn, tp = confusion_matrix(y_true = Y_test, y_pred = Y_predict_test).ravel()
    accuracy = (tn+tp)/(tn+fp+fn+tp)
    precision = tp/(tp+fp)
    recall = tp/(tp+fn)
    f1_score = 2*((precision*recall)/(precision+recall))
    fpr, tpr, thresholds = roc_curve(y_true = Y_test,y_score=Y_predict_test,pos_label=1)
    area_under_curve = auc(fpr, tpr)
    accuracy_list_test.append(accuracy)
    precision_list_test.append(precision)
    recall_list_test.append(recall)
    F1_score_list_test.append(f1_score)
    auc_list_test.append(area_under_curve)
```

```python
from sklearn.cluster import KMeans
from scipy.spatial import distance

accuracy_list_test = []
precision_list_test = []
recall_list_test = []
F1_score_list_test = []
auc_list_test = []
accuracy_list_train = []
precision_list_train = []
recall_list_train = []
F1_score_list_train = []
auc_list_train = []
```

```python
for M in range(0,30):
    rs = M    
    X_B_train, X_B_test, Y_B_train, Y_B_test = model_selection.train_test_split(df_B_X, df_B_Y, train_size=0.8,test_size=0.2, random_state=rs);
    X_M_train, X_M_test, Y_M_train, Y_M_test = model_selection.train_test_split(df_M_X, df_M_Y, train_size=0.8,test_size=0.2, random_state=rs);
    #X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X, Y, train_size=0.7,test_size=0.3, random_state=10);
    X_test = pd.concat([X_B_test,X_M_test])
    Y_test = pd.concat([Y_B_test,Y_M_test])
    X_train = pd.concat([X_B_train,X_M_train])
    Y_train = pd.concat([Y_B_train,Y_M_train])
    
    kmeans = KMeans(n_clusters=2, n_jobs=-1, random_state=rs)
    kmeans.fit(X_train)
    
    # calculate distance between each data to centroid
    distance_list = distance.cdist(X_train, kmeans.cluster_centers_,'euclidean')
    distance_list1 = []
    distance_list2 = []
    for each_distance in distance_list:
        distance_list1.append(each_distance[0])
        distance_list2.append(each_distance[1])
    index_list1 = sorted(range(len(distance_list1)), key=lambda k: distance_list1[k])[:30]
    index_list2 = sorted(range(len(distance_list2)), key=lambda k: distance_list2[k])[:30]
    label_list1 = []
    label_list2 = []
    
    # count majority
    Y_train_list = Y_train['label'].tolist()
    for index in range(0,30):
        current_row = index_list1[index]
        label_list1.append(Y_train_list[current_row])
        current_row = index_list2[index]
        label_list2.append(Y_train_list[current_row])    
    cluster_label1 = get_mode(label_list1)
    cluster_label2 = get_mode(label_list2)
    
    # get predict label for each data
    X_train_labels_list = list(kmeans.labels_)
    Y_predict_train = []
    for each_label in X_train_labels_list:
        if each_label==0:
            Y_predict_train.append(cluster_label1)
        else:
            Y_predict_train.append(cluster_label2)
            
        
    # testing
    distance_list = distance.cdist(X_test, kmeans.cluster_centers_,'euclidean')
    Y_predict_test = []
    for each_distance in distance_list:
        if each_distance[0]<each_distance[1]:
            Y_predict_test.append(cluster_label1)
        else:
            Y_predict_test.append(cluster_label2)
                     
            
    # Train
    tn, fp, fn, tp = confusion_matrix(y_true = Y_train, y_pred = Y_predict_train).ravel()
    accuracy = (tn+tp)/(tn+fp+fn+tp)
    precision = tp/(tp+fp)
    recall = tp/(tp+fn)
    f1_score = 2*((precision*recall)/(precision+recall))
    fpr, tpr, thresholds = roc_curve(y_true = Y_train,y_score=Y_predict_train,pos_label=1)
    area_under_curve = auc(fpr, tpr)
    accuracy_list_train.append(accuracy)
    precision_list_train.append(precision)
    recall_list_train.append(recall)
    F1_score_list_train.append(f1_score)
    auc_list_train.append(area_under_curve)
    
    
    # Test
    tn, fp, fn, tp = confusion_matrix(y_true = Y_test, y_pred = Y_predict_test).ravel()
    accuracy = (tn+tp)/(tn+fp+fn+tp)
    precision = tp/(tp+fp)
    recall = tp/(tp+fn)
    f1_score = 2*((precision*recall)/(precision+recall))
    fpr, tpr, thresholds = roc_curve(y_true = Y_test,y_score=Y_predict_test,pos_label=1)
    area_under_curve = auc(fpr, tpr)
    accuracy_list_test.append(accuracy)
    precision_list_test.append(precision)
    recall_list_test.append(recall)
    F1_score_list_test.append(f1_score)
    auc_list_test.append(area_under_curve)
```

```python
for M in range(0,30):
    rs=M 
    X_B_train, X_B_test, Y_B_train, Y_B_test = model_selection.train_test_split(df_B_X, df_B_Y, train_size=0.8,test_size=0.2, random_state=rs);
    X_M_train, X_M_test, Y_M_train, Y_M_test = model_selection.train_test_split(df_M_X, df_M_Y, train_size=0.8,test_size=0.2, random_state=rs);
    #X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X, Y, train_size=0.7,test_size=0.3, random_state=10);
    X_test = pd.concat([X_B_test,X_M_test])
    Y_test = pd.concat([Y_B_test,Y_M_test])
    X_train = pd.concat([X_B_train,X_M_train])
    Y_train = pd.concat([Y_B_train,Y_M_train])
    
    # normalize data
    min_max_scaler = preprocessing.MinMaxScaler()
    x_scaled = min_max_scaler.fit_transform(X_train)
    X_train_normalize = pd.DataFrame(x_scaled)
    
    # predict data by SpectralClustering
    sc = SpectralClustering(n_clusters=2,affinity ='rbf',gamma=1,n_jobs=-1,random_state=rs)
    Y_train_sc_predict = sc.fit_predict(X_train_normalize)
    
    # count cluster mode label
    Y_train_list = Y_train['label'].tolist()
    list_of_zero = []
    list_of_one = []
    label_list = list(sc.labels_)
    for index in range(0,len(label_list)):
        if label_list[index]==0:
            list_of_zero.append(Y_train_list[index])
        else:
            list_of_one.append(Y_train_list[index])
    cluster_label0 = get_mode(list_of_zero)
    cluster_label1 = get_mode(list_of_one)
    
    # apply KNN with K=1 and predict
    min_max_scaler = preprocessing.MinMaxScaler()
    x_scaled = min_max_scaler.fit_transform(X_test)
    X_test_normalize = pd.DataFrame(x_scaled)

    neigh = KNeighborsClassifier(n_neighbors=1,n_jobs=-1)
    neigh.fit(X_train_normalize, Y_train_sc_predict)
    #Y_predict_train = neigh.predict(X_train_normalize)
    #Y_predict_test = neigh.predict(X_test_normalize)
    temp_train = neigh.predict(X_train_normalize)
    temp_test = neigh.predict(X_test_normalize)
    
    # training
    Y_predict_train = []
    for each_label in temp_train:
        if each_label==0:
            Y_predict_train.append(cluster_label0)
        else:
            Y_predict_train.append(cluster_label1)
    
    # testing
    Y_predict_test = []
    for each_label in temp_test:
        if each_label==0:
            Y_predict_test.append(cluster_label0)
        else:
            Y_predict_test.append(cluster_label1)        
            

    # Train
    tn, fp, fn, tp = confusion_matrix(y_true = Y_train, y_pred = Y_predict_train).ravel()
    accuracy = (tn+tp)/(tn+fp+fn+tp)
    precision = tp/(tp+fp)
    recall = tp/(tp+fn)
    f1_score = 2*((precision*recall)/(precision+recall))
    fpr, tpr, thresholds = roc_curve(y_true = Y_train,y_score=Y_predict_train,pos_label=1)
    area_under_curve = auc(fpr, tpr)
    accuracy_list_train.append(accuracy)
    precision_list_train.append(precision)
    recall_list_train.append(recall)
    F1_score_list_train.append(f1_score)
    auc_list_train.append(area_under_curve)


    # Test
    tn, fp, fn, tp = confusion_matrix(y_true = Y_test, y_pred = Y_predict_test).ravel()
    accuracy = (tn+tp)/(tn+fp+fn+tp)
    precision = tp/(tp+fp)
    recall = tp/(tp+fn)
    f1_score = 2*((precision*recall)/(precision+recall))
    fpr, tpr, thresholds = roc_curve(y_true = Y_test,y_score=Y_predict_test,pos_label=1)
    area_under_curve = auc(fpr, tpr)
    accuracy_list_test.append(accuracy)
    precision_list_test.append(precision)
    recall_list_test.append(recall)
    F1_score_list_test.append(f1_score)
    auc_list_test.append(area_under_curve)
```

```python
for M in range(0,50):
    whole_train = pd.concat([X_train, Y_train], axis=1)
    whole_train = whole_train.sample(frac=1)#.reset_index(drop=True)

    while not contain_both_class(whole_train.head(10)):
        whole_train = whole_train.sample(frac=1)

    current_pool = pd.DataFrame()
    part_train = whole_train.copy()
    test_error_list = []

    #--------------------------------------initialize the linear svm-------------------------------------
    # use CV find best C
    current_pool = current_pool.append(part_train.head(10))
    part_train = part_train.iloc[10:]
    current_pool_X = current_pool.iloc[:,:-1]
    current_pool_Y = current_pool.iloc[:,-1:]
    param_grid = {
        'penalty':['l1'],
        'C':c_list,
        'dual':[False]
    }
    clf = GridSearchCV(estimator = LinearSVC(), n_jobs=-1, param_grid=param_grid,cv=5,scoring='accuracy')
    clf.fit(current_pool_X, current_pool_Y)
    best_C = clf.best_params_.get('C')

    # apply best C to lsvc and predict
    lsvc = LinearSVC(penalty='l1',C=best_C,dual=False)
    lsvc.fit(current_pool_X,current_pool_Y)
    Y_predict_test = lsvc.predict(X_test)
    # calculate test error
    tn, fp, fn, tp = confusion_matrix(Y_test, Y_predict_test).ravel()
    test_error = (fp+fn)/(Y_test.iloc[:,0].size)
    test_error_list.append(test_error)

    for j in range(0,89):
        # add distance col to part_train
        part_train_X = part_train.iloc[:,:4]
        part_train_Y = part_train.iloc[:,4:5]
        y = lsvc.decision_function(part_train_X)
        w_norm = np.linalg.norm(lsvc.coef_)
        distance_list = list(y / w_norm)
        distance_abs_list =  [abs(ele) for ele in distance_list] 
        part_train['distance'] = distance_abs_list 

        # sort part_train based on distance and take head(10) to current_pool
        part_train = part_train.sort_values(by=['distance'])
        part_train = part_train.drop(columns=['distance'])
        current_pool = current_pool.append(part_train.head(10))
        part_train = part_train.iloc[10:]

        # use CV find best C
        current_pool_X = current_pool.iloc[:,:-1]
        current_pool_Y = current_pool.iloc[:,-1:]
        param_grid = {
            'penalty':['l1'],
            'C':c_list,
            'dual':[False]
        }
        clf = GridSearchCV(estimator = LinearSVC(), n_jobs=-1, param_grid=param_grid,cv=5,scoring='accuracy')
        clf.fit(current_pool_X, current_pool_Y)
        best_C = clf.best_params_.get('C')

        # apply best C to lsvc and predict
        lsvc = LinearSVC(penalty='l1',C=best_C,dual=False)
        lsvc.fit(current_pool_X,current_pool_Y)
        Y_predict_test = lsvc.predict(X_test)
        # calculate test error
        tn, fp, fn, tp = confusion_matrix(Y_test, Y_predict_test).ravel()
        test_error = (fp+fn)/(Y_test.iloc[:,0].size)
        test_error_list.append(test_error)
    all_test_error_list_active.append(test_error_list)
```
