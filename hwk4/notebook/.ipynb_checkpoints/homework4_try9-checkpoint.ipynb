```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
import math
from IPython.display import display
import warnings
import statistics
warnings.filterwarnings('ignore')

def get_split(df,n):
    num_of_cols = df.iloc[:,0].size
    n_floor = math.floor(num_of_cols/n)
    remainder = num_of_cols%n
    temp1 = [n_floor]*(n-remainder)
    temp2 = [n_floor+1]*(remainder)
    list_of_num_partition = temp1+temp2
    #print(list_of_num_partition)
    other_df = df.copy()
    part_df_list = []
    for p_rows in list_of_num_partition:
        temp = other_df.head(p_rows)
        other_df = other_df.iloc[p_rows:]
        part_df_list.append(temp)
    
    result = []
    for j in range(0,n):
        train_df = pd.DataFrame()
        for i in range(0,n):
            if i!=j:
                temp = part_df_list[i]
                train_df = pd.concat([train_df,temp])
            else:
                test_df = part_df_list[i]
        result.append([train_df,test_df])        
            
    return result
```

```python
import xgboost as xgb
from sklearn.model_selection import KFold

temp_list = list(range(-5,6))
alpha_list = []
for i in temp_list:
    alpha_list.append(pow(10,i))

kf = KFold(n_splits=10, shuffle=True, random_state=0)
mse_list = []
for current_alpha in alpha_list:
    error_list = []
    for train_index, test_index in kf.split(X=train_data):   
        train_index = list(train_index)
        test_index = list(test_index)
        model = xgb.XGBRegressor(n_jobs=1,reg_alpha=current_alpha)
        model.fit(train_data.iloc[train_index,:], train_target.take(train_index))
        predictions = model.predict(train_data.iloc[test_index,:])
        real = train_target.take(test_index)
        error_list.append(mean_squared_error(real, predictions))
    mse=mean(error_list)
    mse_list.append(mse)
    
index = mse_list.index(min(mse_list))
best_alpha = alpha_list[index]
print('Firstly, we use 10-folds cross validation determine alpha from '+str(alpha_list)+',')
print('The best alpha = '+str(best_alpha))
print('To find potential better alpha, we use cross validation again to find alpha from a smaller increment of list, which is near '+str(best_alpha))
```
