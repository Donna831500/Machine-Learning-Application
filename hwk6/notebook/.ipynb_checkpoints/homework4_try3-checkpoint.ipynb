```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
import math
from IPython.display import display
import warnings
warnings.filterwarnings('ignore')
```

<h5>part (a)<h5>

```python
l1 = list(range(1,21))
l2 = list(range(22,16022))
skip_rows = l1+l2
skip_rows = list(range(0,20))
df = pd.read_csv('../data/communities.data',
                na_values='?',
                index_col=None,
                header = None)

headers = []
for i in range(0,127):
    headers.append('feature_'+str(i))
headers.append('ViolentCrimesPerPop')
df.columns = headers

train_df = df.head(1495)
test_df = df.iloc[1495:]
#train_df.info(verbose=True)
#display(train_df.dtypes)
display(test_df)
#df[1].isna().sum()
#df[0].max()
```

<h5>part (b)<h5>

```python
# ------------------------data imputation-------------------------------
from sklearn.impute import KNNImputer
# train
train_features_df_raw = train_df.iloc[:, 5: 127]
imputer = KNNImputer(n_neighbors=5, weights="uniform")
x = imputer.fit_transform(train_features_df_raw)
train_features_df = pd.DataFrame(x)
train_features_df

# test
test_features_df_raw = test_df.iloc[:, 5: 127]
imputer = KNNImputer(n_neighbors=5, weights="uniform")
x = imputer.fit_transform(test_features_df_raw)
test_features_df = pd.DataFrame(x)
test_features_df
```

<h5>part (c)<h5>

```python
train_features_df.corr()
```

<h5>part (d)<h5>

```python
from scipy.stats import variation
#train_features_df.loc[train_features_df.apply(pd.to_numeric, errors='coerce').isnull().any(1)]
cv = variation(a=train_features_df, axis = 0,nan_policy='omit')
print(cv)
```

<h5>part (e)<h5>

```python
number_of_feature = 11
indices = list(np.argsort(cv)[-11:])
indices_list = []
for i in range(0,11):
    indices_list.append('feature_'+str(indices[i]+5))

df_draw = train_features_df[indices]
df_draw.columns = indices_list
df_draw['ViolentCrimesPerPop']=train_df['ViolentCrimesPerPop']
df_draw

# draw
fig, ((ax1, ax2, ax3), (ax4, ax5, ax6), (ax7, ax8, ax9),(ax10, ax11, ax12)) = plt.subplots(4, 3)

ax_list = [ax1,ax2,ax3,ax4,ax5,ax6,ax7,ax8,ax9,ax10,ax11]
for i in range(0,11):
    ax_list[i].scatter(df_draw[indices_list[i]], df_draw['ViolentCrimesPerPop'], color = 'orange')
    ax_list[i].set_xlabel(indices_list[i])
    ax_list[i].set_ylabel('ViolentCrimesPerPop')

fig.set_figheight(20)
fig.set_figwidth(15)
```

```python
temp = df_draw.iloc[:, 0: 11]
seaborn.boxplot(data = temp)
seaborn.set(rc={'figure.figsize':(20,20)})
```

<h5>part (f)<h5>

```python
import statsmodels.formula.api as sm
from sklearn.metrics import mean_squared_error
# build model
f = ''
for i in range(0,11):
    f = f+indices_list[i]
    f = f+'+'
f = f[:-1]
f = 'ViolentCrimesPerPop~'+f
result = sm.ols(formula=f, data=df_draw).fit()
print(result.summary())

# prediction
test_df_data = test_features_df[indices]
test_df_data.columns = indices_list
test_df_label = test_df['ViolentCrimesPerPop']
predictions = result.predict(test_df_data)
#print(predictions)
test_error = mean_squared_error(test_df_label, predictions)
test_error
```

```python
from sklearn import linear_model
temp = df_draw.iloc[:, 0: 11]
temp1 = df_draw.iloc[:, 11]

lm = linear_model.LinearRegression()
model = lm.fit(X=temp, y=temp1)
predictions = lm.predict(test_df_data)
print(predictions)[0:5]
```
