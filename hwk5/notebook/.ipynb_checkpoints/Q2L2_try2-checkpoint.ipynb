```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix,roc_curve,auc
import math
from IPython.display import display
import warnings
import statistics
warnings.filterwarnings('ignore')
from sklearn.impute import KNNImputer
from scipy.stats import variation
from sklearn.metrics import mean_squared_error
from sklearn import linear_model
from sklearn.linear_model import RidgeCV
from sklearn.linear_model import LassoCV
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_validate
from sklearn.cross_decomposition import PLSRegression
from statistics import mean
import xgboost as xgb
from sklearn.model_selection import KFold

def convert_label(s):
    if s=='neg':
        return 0;
    elif s=='pos':
        return 1;
```

```python
#------------------------------part b)----------------------------------
l1 = list(range(0,20))

# read train data
train_df_raw = pd.read_csv('../data/aps_failure_training_set.csv',
                na_values='na',
                index_col=None,
                skiprows=l1)
all_columns_list = list(train_df_raw.columns)

# read test data
test_df_raw = pd.read_csv('../data/aps_failure_test_set.csv',
                na_values='na',
                index_col=None,
                skiprows=l1)
```

```python
#  i)
from sklearn.impute import SimpleImputer
print('Mean,mode and knn value can be fitted in to missing values. I use mean in this case.')

# train
train_features_df_raw = train_df_raw.iloc[:, 1:]
test_features_df_raw = test_df_raw.iloc[:, 1:]
whole_features_df_raw = pd.concat([train_features_df_raw,test_features_df_raw])
whole_df_raw = pd.concat([train_df_raw,test_df_raw])
train_size = train_features_df_raw.iloc[:,0].size


imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(whole_features_df_raw)
x = imputer.transform(whole_features_df_raw)
whole_features_df = pd.DataFrame(x)
train_features_df = whole_features_df.head(train_size)
test_features_df = whole_features_df.iloc[train_size:]
```

```python
from sklearn.ensemble import RandomForestClassifier
# impute test data
train_data = train_features_df.copy()
train_target = train_df_raw['class'].apply(convert_label)
test_data = test_features_df.copy()
test_target = test_df_raw['class'].apply(convert_label)

whole_train_data = train_data.copy()
whole_train_data['label']=train_target
whole_test_data = test_data.copy()
whole_test_data['label']=list(test_target)

whole_test_data
```

```python
# train_data
# train_target
# test_data
# test_target
#------------------------------part c)----------------------------------
from sklearn.ensemble import RandomForestClassifier
# impute test data
train_data = train_features_df.copy()
train_target = train_df_raw['class'].apply(convert_label)
test_data = test_features_df.copy()
test_target = test_df_raw['class'].apply(convert_label)

whole_train_data = train_data.copy()
whole_train_data['label']=list(train_target)
whole_test_data = test_data.copy()
whole_test_data['label']=list(test_target)

train_pos=whole_train_data[whole_train_data['label']==1]
train_neg=whole_test_data[whole_test_data['label']==0]

ds_train_pos=train_pos.head(100)
ds_train_neg=train_neg.head(5900)

ds_train = pd.concat([ds_train_pos,ds_train_neg])
ds_train
```

```python
#------------------------------part e)----------------------------------
import weka.core.jvm as jvm
from weka.classifiers import Classifier, Evaluation
from weka.core.converters import Loader
from weka.core.classes import Random
from weka.core.dataset import create_instances_from_matrices
import numpy as np
```

```python
jvm.start(packages=True)
```

```python
# Load data
#x = np.array(train_data)
#y = np.array(train_target)
#train_dataset = create_instances_from_matrices(x, y,name="training dataset")

x = np.array(ds_train)
train_dataset = create_instances_from_matrices(x, name="training dataset")
x = np.array(test_data)
y = np.array(test_target)
test_dataset = create_instances_from_matrices(x, y,name="test dataset")
```

```python
from weka.filters import Filter
train_dataset.class_is_last()
test_dataset.class_is_last()
f = Filter(classname="weka.filters.unsupervised.attribute.NumericToNominal", options=['-R', '171'])
f.inputformat(test_dataset)
train_dataset = f.filter(train_dataset)
test_dataset = f.filter(test_dataset)
```

```python
# apply tree
cls = Classifier(classname="weka.classifiers.trees.LMT")
cls.build_classifier(train_dataset)
```

```python
from weka.classifiers import Evaluation
from weka.core.classes import Random
evl = Evaluation(train_dataset)
```

```python
evl.crossvalidate_model(cls, train_dataset, 10, Random(1))
```

```python
print(evl.percent_correct)
print(evl.summary())
print(evl.class_details())
```

```python
import weka.plot.classifiers as plcls  # NB: matplotlib is required
plcls.plot_roc(evl, class_index=[0, 1], wait=True)
```

```python
print(evl.confusion_matrix)
```
